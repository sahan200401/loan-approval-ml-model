```python
import plotly.express as px

missing = df.isnull().sum().reset_index()
missing.columns = ['column','Missing values']

fig = px.bar(
    missing,
    x='column',
    y='Missing values',
    title='Missing values for each column'
)

fig.show()
```

```python
print("filling missing values of age")

df['age'] = df['age'].apply(lambda x : np.nan if x > 100 else x)
df['age'].fillna(df['age'].median(),inplace = True)
```

```python
df['loan_interest_rate'].fillna(df['loan_interest_rate'].median(), inplace=True)
mode_val = df['payment_default_on_file'].mode()[0]
df['payment_default_on_file'].fillna(mode_val, inplace=True)
```
